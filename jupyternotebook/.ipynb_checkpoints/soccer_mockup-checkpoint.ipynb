```python
# encoding = utf-8

import pandas as pd
mockup = pd.read_csv('2018_mockup.csv')
mockup['日期'] = pd.to_datetime(mockup['日期'])
print(mockup[0:10])
```

```python
import matplotlib.pyplot as plt
nadropped_mockup=mockup.dropna(axis = 0)
plt.plot(nadropped_mockup['日期'],nadropped_mockup['累计盈亏'])
plt.xticks(rotation=45)
plt.show()
```
