```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot(flip = 1):
    x = np.linspace(0,14,100)
    #find 100 points in 0-14 
    for i in range(1,7):
        plt.plot(x, np.sin(x + i*.5)*(7-i)*flip)

sinplot()
    
```

```python
sns.set()
sinplot()
```
