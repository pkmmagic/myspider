```python
import pandas as pd
reviews = pd.read_csv('fandango_score_comparison.csv')
cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue','Fandango_Stars']
norm_reviews = reviews[cols]

print(norm_reviews[:1])

```

```python
import matplotlib.pyplot as plt
from numpy import arange

num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue','Fandango_Stars']
agencies = ['Rotten', 'Metacritic', 'IMDB', 'Fandango','Fandango S']
bar_heights = norm_reviews.loc[0, num_cols].values
#print(bar_heights)
bar_positions = arange(5) + 0.75
tick_positions = range(1,6)
fig, ax = plt.subplots()


ax.set_xticks(tick_positions)
ax.set_xticklabels(agencies, rotation = 45)
ax.bar(bar_positions, bar_heights, 0.3)
#0.3 = column width

ax.set_xlabel('Agencies')
ax.set_ylabel('Average Ratings')
ax.set_title('Average Ratings for Avengers: Age of Ultron (2015)')
plt.show()

```

```python
import matplotlib.pyplot as plt
from numpy import arange

num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue','Fandango_Stars']
agencies = ['Rotten', 'Metacritic', 'IMDB', 'Fandango','Fandango S']
bar_widths = norm_reviews.loc[0, num_cols].values
#print(bar_heights)
bar_positions = arange(5) + 1.1
tick_positions = [5,4,3,2,1]
fig, ax = plt.subplots()


ax.set_yticks(tick_positions)
ax.set_yticklabels(agencies, rotation = 45)
ax.barh(bar_positions, bar_widths, 0.3)
#0.3 = column width

ax.set_ylabel('Agencies')
ax.set_xlabel('Average Ratings')
ax.set_title('Average Ratings for Avengers: Age of Ultron (2015)')
plt.show()
```
