```python
# encoding = utf-8

import pandas as pd
import matplotlib.pyplot as plt
mockup = pd.read_csv('2014_mockup.csv')
mockup['日期'] = pd.to_datetime(mockup['日期'])


nadropped_mockup=mockup.dropna(axis = 0)

print(mockup)

```

```python
#print(nadropped_mockup[0:10]['盘口'])
PKadvised = nadropped_mockup['盘口'].str.split('/')
PKadvised.index=range(len(PKadvised))
SCOREadvised = nadropped_mockup['比分'].str.split('-')
SCOREadvised.index=range(len(SCOREadvised))

for i in range(len(SCOREadvised)):
    if type(SCOREadvised[i])==list:
        r = float(SCOREadvised[i][0]) - float(SCOREadvised[i][1])
        SCOREadvised[i]=r


print(SCOREadvised)
#print(type(advised[1]))




```

```python

for i in range(len(PKadvised)):
    if len(PKadvised[i])>1:
        r = float(PKadvised[i][0]) + float(PKadvised[i][1])
        PKadvised[i]=r/2
    else:
        PKadvised[i] = float(PKadvised[i][0])
    
print(PKadvised)
```

```python
plt.scatter(PKadvised,SCOREadvised)

plt.show()
```
