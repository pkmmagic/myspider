```python


import pandas as pd
unrate = pd.read_csv('unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
print(unrate)
                                
```

```python
import matplotlib.pyplot as plt
nt48 = unrate[0:12]
plt.plot(nt48['DATE'],nt48['VALUE'])
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Rate of US in 1948')
plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month
fig = plt.figure(figsize=(10,3))

plt.plot(unrate[0:6]['MONTH'],unrate[0:6]['VALUE'],c='red',label='Jan-Jun')
plt.plot(unrate[6:12]['MONTH'],unrate[6:12]['VALUE'],c='blue',label='Jul-Dec')

plt.legend(loc=0)

print(help(plt.legend))
plt.show()
```
